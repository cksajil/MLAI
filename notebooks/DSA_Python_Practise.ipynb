# Loops

## For Loop

```python
word = "data"
for char in word:
    print(char)
```

```python
word = "data"
for char in word[::-1]:
    print(char)
```

```python
for i in range(1,10,3):
    print(i)
```

```python
numbers = [1,2,3,4,5]
for i in numbers:
    if i == 3:
        break
    print(i)
print("outside of for loop ")
```

```python
numbers = [1,2,3,4,5]
for i in numbers:
    if i == 3:
        continue
    print(i)
print("outside of for loop ")
```
