```python
plt.hist(data["mpg"], color="green", rwidth=0.8)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Count")
plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks()
plt.show()
```

```python
plt.figure(figsize=(5, 2))
plt.scatter(data["weight"], data["mpg"], c="k", marker="^")
plt.title("MPG vs Weight")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.scatter(data["weight"], data["mpg"], c="k")
plt.ylabel("MPG")
plt.xlabel("Weight")
plt.subplot(2, 1, 2)
plt.scatter(data["displacement"], data["mpg"], c="blue")
plt.xlabel("Displacement")
plt.ylabel("MPG")
plt.show()
```
