## For Loop

The for loop in Python is used to iterate over a sequence (list, tuple, string) or other iterable objects

```python
for variable in iterable:
  pass
```

```python
from time import sleep
for i in range(1, 11):
  sleep(0.1)
  print('*', end="")
```

```python
from tqdm import tqdm
for i in tqdm(range(1,100)):
  sleep(0.1)
```

## BREAK and CONTINUE

### Break

```python
numbers = range(10)

for num in numbers:
    if num == 3:
        break
    print(num)
print("outside of for loop ")
```

```python
for num in numbers:
    if num == 3:
        continue
    print(num)
print("outside of for loop ")
```

## Functions


*   Function is a group of related statements that perform a specific task
*   There are bulit-in functions and user defined functions
*   Functions break the program into smaller and modular chunks
*   As program grows, function make it more organised and managable
*   It avoids repetition and make code reusable
*   Helps in efficient utilization of RAM



```python
def function_name(parameters):
  """ doc string"""
  # Statements
  return value
```

```python
print(function_name.__doc__) # prints the docstring of the function
```

```python
def add_numbers(a, b):
  """ This function adds two numbers """
  sum = a + b
  return sum
```

```python
add_numbers(4, 6)
```

```python
print(add_numbers.__doc__)
```

```python
def simple_arithmetic(a,b):
  return a+b, a-b, a*b, a/b

tsum, tdiff, tmul, tdiv = simple_arithmetic(10,5)
print(tsum, tdiff, tmul, tdiv)
```
