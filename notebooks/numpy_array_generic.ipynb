```python
x = np.arange(20)
print(x[:3])
```

```python
mean_x = np.mean(x)
median_x = np.median(x)
std_x = np.std(x)
var_x = np.var(x)
max_x = np.max(x)
min_x = np.min(x)
print("Mean of x is:", mean_x)
print("Median of x is:", median_x)
print("Max of x is:", max_x)
print("Min of x is:", min_x)
print("Standard deviation of x is:", std_x)
print("Variance of x is:", var_x)
```
