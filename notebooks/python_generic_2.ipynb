```python
timestamp = "Wed Jul  12 10:33:55 AM UTC 2025"
# Output: Wednessday 2nd July 10:33 AM, 2025
```

```python
postfix = ["st", "nd","rd", "th"]
weeks = {"Sun": "Sunday",
         "Mon": "Monday",
         "Tue": "Tuesday",
         "Wed": "Wednessday",
         "Fri": "Friday",
         "Sat": "Saturday"}
week, month,_, day, time, m, _, year = timestamp.split(" ")
time = time.split(":")
time = ":".join(time[:2])
outstring = f"{weeks[week]} {day}{postfix[int(day[-1])-1]} {month} {time} {m}, {year}"
print(outstring)
```

```python
# Input: 7458
# Output: Seven Four Five Eight

digits = ["Zero", "One", "Two", "Three", "Four",\
          "Five","Six", "Seven", "Eight", "Nine"]
x = input("Enter the OTP: ")
x = list(map(int, x))
digits_str = [digits[d] for d in x]
print(" ".join(digits_str))
```
