# Basic class definition

## Class attributes

```python
class student:
    def __init__(self, x, y):
        self.name = x
        self.age = y
```

```python
person_1 = student("Bob", 15)
```

```python
print(person_1.name)
print(person_1.age)
```

```python
person_2 = student("Charlie", 16)
```

```python
print(person_2.name)
print(person_2.age)
```

## Class functions

```python
class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def say_name(self):
        print("Hi, my name is, ", self.name)
```

```python
person_1 = student("Bob", 15)
```

```python
person_1.say_name()
```

```python
person_2 = student("Charlie", 16,)
```

```python
person_2.say_name()
```

## Inheritance

```python
class Animal:
    def __init__(self, name, age, sound):
        self.name = name
        self.age = age
        self.sound = sound

    def make_sound(self):
        print(self.sound)
```

```python
class Dog(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)

    def action(self):
        print("Moves tails")


class Cat(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)

    def action(self):
        print("Moves ears")
```

```python
my_dog = Dog("Buddy", 3, "Woof!")
```

```python
my_dog.make_sound()
```

```python
my_dog.action()
```

```python
my_cat = Cat("Whiskers", 2, "Meaw!!")
```

```python
my_cat.make_sound()
```

```python
my_cat.action()
```
