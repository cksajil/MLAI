```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
print(digits['data'].shape)
print(digits['target'].shape)
```

```python
tsne = TSNE(n_components=2, random_state=42)
```

```python
x_transformed = tsne.fit_transform(digits['data'])
```

```python
tsne_df = pd.DataFrame(np.column_stack((x_transformed, digits['target'])), columns=['X', 'Y', "Targets"])
tsne_df.loc[:, "Targets"] = tsne_df.Targets.astype(int)
```

```python
tsne_df.head(10)
```

```python
plt.figure(figsize=(6,6))
g = sns.FacetGrid(data=tsne_df, hue='Targets', height=8)
g.map(plt.scatter, 'X', 'Y').add_legend()
plt.show()
```
