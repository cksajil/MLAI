```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
# Instantiate the SFS object
# k_features=2: selects 2 features
# forward=True: performs forward selection
# scoring='accuracy': evaluates performance based on accuracy
# cv=5: uses 5-fold cross-validation
sfs1 = SFS(knn,
           k_features=2,
           forward=True,
           floating=False,
           verbose=2,
           scoring='accuracy',
           cv=5)


sfs1 = sfs1.fit(X, y)
```

```python
selected_features = sfs1.k_feature_idx_
print('Selected features:', selected_features)
```

```python
performance = sfs1.k_score_
print('Performance with selected features:', performance)
```
