```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data1 = data.drop('User ID', axis=1)
```

```python
data1.head()
```

```python
data1 = pd.get_dummies(data1)
```

```python
data1.head()
```

```python
data1['Purchased'].value_counts()
```

```python
X = data1.drop('Purchased', axis=1)
y = data1['Purchased']
```

#### Hold Out Technique

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,
                                                   random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LRmodel = LogisticRegression()
LRmodel = LRmodel.fit(X_train, y_train)
```

```python
score_LR = LRmodel.score(X_test, y_test)
```

```python
score_LR
```

#### K-fold Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
kfold_validator = KFold(10)
```

```python
for train_index, test_index in kfold_validator.split(X, y):
    print('Training Index: ', train_index)
    print('Test Index: ', test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result = cross_val_score(LRmodel, X, y, cv=kfold_validator)
```

```python
cv_result
```

```python
np.mean(cv_result)
```

#### Stratified K-Fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator = StratifiedKFold(n_splits=10)
```

```python
for train_index, test_index in skfold_validator.split(X, y):
    print('Training Index: ', train_index)
    print('Test Index: ', test_index)
```

```python
skcv_result = cross_val_score(LRmodel, X, y, cv=skfold_validator)
```

```python
skcv_result
```

```python
np.mean(skcv_result)
```

# Feature Engineering

```python
df = pd.read_csv('demand-forecast.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df['year'] = df['date'].apply(lambda x:x.year)
```

```python
df.head()
```

```python
df['month'] = df['date'].apply(lambda x:x.month)
df['day'] = df['date'].apply(lambda x:x.day)
```

```python
df.head()
```

```python
df['dayOfWeek'] = df['date'].apply(lambda x:x.dayofweek)
```

```python
df['quarter'] = df['date'].apply(lambda x:x.quarter)
```

```python
df.head()
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
def timeOfDay(n):
    if n in range(1, 4):
        return 'Late Night'
    elif n in range(4, 7):
        return 'Early Morning'
    elif n in range(7, 12):
        return 'Morning'
    elif n in range(12, 15):
        return 'Afternoon'
    elif n in range(15, 18):
        return 'Evening'
    elif n in range(18, 24) or n==0:
        return 'Night'
```

```python
df['typeOfDay'] = df['hour'].apply(lambda x:timeOfDay(x))
```

```python
df.head()
```

```python
def monthToSeasons(x):
    if x in [9,10,11]:
        return 'Spring'
    elif x in [12, 1, 2]:
        return 'Winter'
    elif x in [3, 4, 5]:
        return 'Summer'
    elif x in [6, 7, 8]:
        return 'Autumn'
```

```python
df['Season'] = df['month'].apply(monthToSeasons)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df['Season'], y=df['demand'])
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df['dayOfWeek'], y=df['demand'])
```
