# Conditional Statements In Python

## if

if statement: This checks a condition, and if it's true, the indented code block under it will be executed.

```python
a = 2
b = 1
if b>a:
   print("b is greater than a")
   print("inside if block")
print("outside if block")
```

```python
a = 20
b = 30
if None:
    print("Inside if block")
else:
    print("Inside else block")
```

None, False, 0 are interpreted as False. All others are equivaluent to True in if-else conditions

```python
a = "0"
if a:
    print("Inside if block")
else:
    print("Inside Else block")
```

## elif
elif statement: Short for "else if," this allows you to check multiple conditions. It comes after an if statement and only executes if the previous if condition was false, and the elif condition is true.

```python
x = 8
if x > 10:
    print("x is greater than 10")
elif x == 10:
    print("x is exactly 10")
elif x < 10:
    print("x is less than 10")
```

## else
else statement: This captures any remaining cases that donâ€™t match the if or elif conditions. It executes if all previous conditions are false.

```python
x = 10
if x > 10:
    print("x is greater than 10")
elif x == 10:
    print("x is exactly 10")
else:
    print("x is less than 10")
```

## Nested Conditional Statements
You can also nest conditional statements for more complex checks:

```python
digits = {0:'zero', 1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight', 9:'nine'}
x = input("Enter your input: ").
for digit in x:
        print(digits[int(digit)], end=" ")
```

```python
x = 15
if x > 10:
    print("x is greater than 10")
    if x > 20:
        print("x is also greater than 20")
    else:
        print("x is not greater than 20")
```

# Q7. Write code to print a given 4 digit OTP number in text form (e.g. 0123-> Zero One Two Three)

```python
digits = ["Zero", "One", "Two", "Three", "Four",\
          "Five", "Six", "Seven", "Eight", "Nine"]
n = input("Enter a number: ")
a,b,c,d = int(n[0]),int(n[1]),int(n[2]),int(n[3])
print(digits[a], digits[b],digits[c], digits[d])
```
