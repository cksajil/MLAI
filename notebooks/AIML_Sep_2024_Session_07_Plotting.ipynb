## Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 2, 22, 22, 10]
plt.plot(x, y)
plt.xlim(1, 5)
# plt.ylim(-0.5, 0.5)
plt.xlabel("days")
plt.ylabel("counts")
plt.show()
```

### Line plot Axis controls

```python
import numpy as np
f = 50
x = np.arange(0, 400)
y = np.cos(2*np.pi*f*x/2000)
plt.plot(x, y)
# plt.xlim(300, 400)
# plt.ylim(-0.5, 0.5)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

### Multi-line line plots


*   Labels and Legends



```python
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 3, 2, 22, 22, 10]
y2 = [1, 7, 5, 4, 3, 8]
plt.plot(x, y1, c="g", label="location1")
plt.plot(x, y2, c="r", label="location2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend()
plt.show()
# Colors: r:red, g:green,b:blue,m:magenta,c:cyan,k:black
```

## Bar Charts

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
plt.bar(x, item_1_sales, label="item1")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
item_2_sales = [3, 4, 5, 6, 5]
plt.bar(x, item_1_sales, color="r", label="item1")
plt.bar(x, item_2_sales, color="b", alpha=0.2,label="item2")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.legend()
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

## Pie Charts

```python
countries = ["India", "USA", "UK", "UAE"]
population = [123, 20, 30, 40]
plt.pie(population, labels=countries, autopct="%1.2f%%")
plt.show()
```

## Scatter Plot

```python
import pandas as pd
data = pd.read_csv("auto-mpg.csv")
plt.figure(figsize=(6, 3.5))
plt.scatter(data["weight"], data["mpg"], c="m", marker="o")
plt.title("MPG vs Weight")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

## Sub-Figures

```python
plt.figure(figsize=(5, 12))
plt.subplot(2, 1, 1)
plt.scatter(data["weight"], data["mpg"], c="k")
plt.ylabel("MPG")
plt.xlabel("Weight")
plt.subplot(2, 1, 2)
plt.hist(data["displacement"])
plt.xlabel("Displacement")
plt.ylabel("MPG")
plt.show()
```

## Box Plot

```python
a = np.random.normal(170, 1, 20000)
q1 = np.quantile(a, 0.25)
q3 = np.quantile(a, 0.75)
iqr = q3-q1
lower_bound = q1-(1.5*iqr)
upper_bound = q3+(1.5*iqr)
sns.boxplot(a)
```
