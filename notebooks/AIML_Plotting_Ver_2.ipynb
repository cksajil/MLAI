## Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 2, 22, 22, 10]
plt.plot(x, y)
plt.xlim(1, 5)
plt.ylim(0, 15)
plt.xlabel("days")
plt.ylabel("counts")
plt.show()
```

### Line plot Axis controls

```python
import numpy as np
f = 50
x = np.arange(0, 400)
y = np.cos(2*np.pi*f*x/2000)
plt.plot(x, y)
plt.xlim(300, 400)
# plt.ylim(-0.5, 0.5)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

### Multi-line line plots


*   Labels and Legends



```python
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 3, 2, 22, 22, 10]
y2 = [1, 7, 5, 4, 3, 8]
plt.plot(x, y1, c="g", label="location1")
plt.plot(x, y2, c="r", label="location2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend()
plt.show()
# Colors: r:red, g:green, b:blue, m:magenta, c:cyan, k:black
```

## Bar Charts

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
plt.bar(x, item_1_sales, label="item1")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
item_2_sales = [3, 4, 5, 6, 5]
plt.bar(x, item_1_sales, color="r", label="item1")
plt.bar(x, item_2_sales, color="b", alpha=0.2,label="item2")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.legend()
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

## Trainguar Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")
data_new = data.drop(["car name", "horsepower"], axis=1)
corr_mat = data_new.corr()
mask = np.triu(np.ones_like(corr_mat, dtype=bool))
sns.heatmap(corr_mat, mask=mask, annot=True, cmap='viridis')
plt.xticks(rotation=45)
plt.show()
```

## Annotating with equations

```python
import numpy as np
f = 10
x = np.arange(0, 400)
y1 = np.cos(2*np.pi*f*x/2000)
y2 = np.sin(2*np.pi*f*x/2000)
plt.plot(x, y1, c="k", label='$y=cos(x)$')
plt.plot(x, y2, c="b", alpha=0.8, label='$y=sin(x)$')
plt.legend(loc='upper right')
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.text(100, 1, "some waves")
plt.grid()
plt.show()
```

## Multiple Axis Graphs

```python
fig1, ax1 = plt.subplots()
x = np.arange(0, 400)
y1 = 100*np.cos(2*np.pi*f*x/2000)
y2 = np.sin(2*np.pi*f*x/2000)
ax2 = ax1.twinx()
ax1.plot(x, y1, label='Signal 1', c='k', linestyle='--')
ax2.plot(x, y2, label='Signal 2', c='b')
ax1.legend()
ax2.legend()
plt.show()
```
