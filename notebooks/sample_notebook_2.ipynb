```python
class Dog:
    def __init__(self, name, age, sound):
        self.name = name
        self.age = age
        self.sound = sound
    def make_sound(self):
        print(self.sound)
```

```python
my_pet = Dog("Bob", 5, "Bow! Bow!")
my_pet2 = Dog("Alice", 3, "Woof! Woof!")
```

```python
my_pet.make_sound()
my_pet2.make_sound()
```
