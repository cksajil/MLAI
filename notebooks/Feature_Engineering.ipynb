#### BoW Example

```python
documents = ["This is the first document.",
             "This document is the second document.",
             "And this is the third one."]

vocabulary = set()
for document in documents:
  for word in document.lower().split():
    vocabulary.add(word)

bow_representation = []
for document in documents:
  document_bow = []
  for word in vocabulary:
    if word in document.lower().split():
      document_bow.append(1)
    else:
      document_bow.append(0)
  bow_representation.append(document_bow)

print("Vocabulary:", list(vocabulary))
for i, doc_bow in enumerate(bow_representation):
  print(f"Document {i+1}:", doc_bow)
```

### Example with Audio Data

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
audio_data, sr = librosa.load("dtmf_sample.wav",
                              sr=None,
                              duration=1)
ipd.Audio(audio_data, rate=sr)
```

### Raw Data

```python
audio_data.shape
```

```python
plt.figure(figsize=(6, 4))
librosa.display.waveshow(audio_data, sr=sr)
# librosa.display.waveshow(audio_data[2400:2500], sr=sr)
plt.title("Time Domain Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

### FFT Spectrum

```python
fft_spectrum = np.fft.fft(audio_data)
frequencies = np.fft.fftfreq(len(audio_data), 1/sr)
plt.figure(figsize=(6, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_spectrum[:len(frequencies)//2]))
plt.title("Frequency Domain Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

### STFT Magnitude

```python
stft_spectrogram = librosa.stft(audio_data)
stft_magnitude = np.abs(stft_spectrogram)
stft_db = librosa.amplitude_to_db(stft_magnitude, ref=np.max)
plt.figure(figsize=(6, 4))
librosa.display.specshow(stft_db, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title("STFT Spectrogram")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.show()
```
