## Min-Max Scaling

```r
data <- read.csv("auto-mpg.csv")
```

```r
head(data)
```

```r
displacement_original <- data$displacement
plot(displacement_original, type = 'l')
```

```r
d_min = min(displacement_original)
d_max  = max(displacement_original)
displacement_minmax = (displacement_original-d_min)/(d_max-d_min)
```

```r
plot(displacement_minmax, type = 'l', col = 'blue')
```

```r
plot(displacement_original, type = 'l', col = 'red')
```

## Standard Scaling

```r
d_std = sd(displacement_original)
displacement_std = (displacement_original-mean(displacement_original))/d_std
plot(displacement_std, type = 'l', col = 'green')
```
