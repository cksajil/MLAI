```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('spam.csv', encoding='Windows-1252')
data.head(9)
```

```python
# remove the unnamed columns
```

```python
messages = data[['v1','v2']]
messages.columns = ["label", "text"]
```

```python
# to vie w the complete data in columns
pd.set_option('display.max_colwidth', None)

messages.head(4)
```

```python
# convert labels to 0 and 1, 1 for ham and 0 for spam
messages['label']=messages['label'].map({'ham':1,'spam':0})
```

```python
messages.head()
```

```python
messages.shape
```

```python
from keras.preprocessing import text
tokenizer = text.Tokenizer()
# tokenizer from text librbary
# tokenizer converts words to integer values
# passint the text as inputs to tokenizer

tokenizer.fit_on_texts(list(messages['text']))

# passing these sms one by one

tokenized_texts = tokenizer.texts_to_sequences(messages['text'])
# converted to ingeer values
```

```python
tokenized_texts[0]
```

```python
len(messages['text'][0])
```

```python
print(len(tokenized_texts[0]))
print(len(tokenized_texts[1]))
```

```python
# to avoid the problem with different sizes
from keras.utils import pad_sequences
#going to pad the tokenizer text
X = pad_sequences(tokenized_texts, maxlen=100)
# it just add some zeros, and amke all sms to the length of 100
```

```python
tokenized_texts[1]
```

```python
X[1]
```

```python
#to get all the words
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
#splitting the data to X and y
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, messages['label'].values, test_size = 0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, SimpleRNN, Dropout
```

```python
model = Sequential()
# embed the data, (convert them to a vector) before passing it

# https://keras.io/api/layers/core_layers/embedding/
# https://keras.io/api/layers/recurrent_layers/simple_rnn/


model.add(Embedding(input_dim = len(tokenizer.word_index)+1, output_dim = 128,input_length=100))
model.add(SimpleRNN(10))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# try it with single dropout first, if overfitting comes up add one more drop out
# take output_dim = other values, 50 to 300 (often range)
```

```python
model.compile(loss ='binary_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
model.summary()
```

```python
# 1141888/128 = 8921
```

```python
model.fit(X_train,y_train,epochs = 10, validation_split = 0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[0]
```

```python
#converting to 0 and 1

y_pred[y_pred>0.5] = 1
y_pred[y_pred<=0.5] = 0
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```
